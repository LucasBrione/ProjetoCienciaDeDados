Dados do CENIPA apontam que de todas as ocorrencias registradas, 1694 foram acidentes, 2887 foram incidentes e 661 foram incidentes graves. Segue abaixo a diferença entre eles.

**Acidente**
Qualquer ocorrencia onde haja intenção de voo e ocorra:
1.   Lesão grave ou morte de qualquer pessoa
2.   Aeronave sofra danos ou falhas estruturais irreparáveis
3.   Aeronave seja considerada desaparecida ou em local inacessível

**Incidente**
É toda ocorrencia que não seja classificada como acidente, associada a operação de uma aeronave, havendo intenção de voo, que afete ou possa afetar a segurança da operação e que gere apenas danos reparáveis.

**Incidente Grave**
É o acidente ocorrido com intenção de voo, sob circunstancias em que um acidente quase ocorreu. A diferença entre o incidente grave e o acidente está apenas nas consequencias.
